```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 绘图中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
"""
data_import
"""
data = pd.read_excel('入境.xls',sheet_name='Data')
# 删除 Indicator Names 列
# 给所有列取名
data.columns = ['国家名称','国家缩写','指标名称','指标代码','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004',
                '2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
data
```

```python
"""
将每一行中缺失值的个数大于10的行删除
"""
data.dropna(thresh=25,inplace=True)
data
```

```python
"""
删除国家名称为缺失值的行
"""
data.dropna(subset=['国家名称'],inplace=True)
data
```

```python
"""
将国家名称这一列作为列标签,进行转置
"""
data = data.set_index('国家名称').T
# 删除国家缩写,指标名称,指标代码这些行
data.drop(['国家缩写','指标名称','指标代码'],inplace=True)
data
```

```python
"""
用平均值填充缺失值
"""
data = data.fillna(data.mean())
data
```

```python
"""
将data中的每一列的数据进行求和
"""
# 各个国家的旅游总人数
country_sum = data.apply(lambda x:x.sum(),axis=0)
# 进行排序
country_sum = country_sum.sort_values(ascending=False)
country_sum
```

```python
"""
将contry_sum导出为excel
"""
country_sum.to_excel('各个国家的旅游总人数.xlsx')
```

这个表格代表了每个国家,地区,地区组织的旅游总人数统计
通过这个表格可以分析每个国家的入境人数,从而分析出哪些国家的旅游人数最多,哪些国家的旅游人数最少

```python
"""
对离境表格进行处理
"""
data_out = pd.read_excel('离境.xls',sheet_name='Data')
data_out.columns = ['国家名称','国家缩写','指标名称','指标代码','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004',
                '2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
data_out.dropna(thresh=25,inplace=True)
data_out.dropna(subset=['国家名称'],inplace=True)
data_out = data_out.set_index('国家名称').T
# 删除国家缩写,指标名称,指标代码这些行
data_out.drop(['国家缩写','指标名称','指标代码'],inplace=True)
# 导出excel
data_out.to_excel('python处理后离境未插值.xlsx')
data_out = data_out.fillna(data.mean())
# 各个国家的旅游总人数
country_sum_out = data_out.apply(lambda x:x.sum(),axis=0)
# 进行排序
country_sum_out = country_sum_out.sort_values(ascending=False)
country_sum_out
```
