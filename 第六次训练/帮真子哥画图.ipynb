```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# 设置画布大小
plt.figure(figsize=(6,4))
# 设置中文字体和负号显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
X1 = [1995,2000,2003,2006,2009,2012,2015,2019]
Y1 = [9374300,10269800,9481400,10285600,12569800,15643800,16254900,23944700]
# 函数1
x = np.linspace(1995,2020)
f1 = 555556*x-1101731890
# 画出函数1和X1,Y1
plt.plot(x,f1)
plt.scatter(X1,Y1)
plt.xlabel('year',fontdict={'size':14})
plt.ylabel('USA_2_GAB',fontdict={'size':14})
plt.legend(['拟合曲线','USA_2_FRA'])
plt.show()
```
